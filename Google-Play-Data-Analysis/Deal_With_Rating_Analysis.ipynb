## Group By in pandas

```python
import pandas as pd
#import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore.csv')

impute = SimpleImputer(missing_values = np.nan , strategy = 'mean')
impute.fit(df.iloc[:,2:3].values)
df.iloc[:,2:3] = impute.transform(df.iloc[:,2:3].values)

df = df.dropna()
```

## Names the apps that are in ART_AND_DESIGN with rating > 4.5 in ascending order

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
df_pr = df_pr[df_pr['Rating'] > 4.5]

df_pr.sort_values(by = 'Rating' , ascending = True).head(10)
```

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
df_pr = df_pr[df_pr['Rating'] > 4.5]

df_pr.sort_values(by = 'Rating' , ascending = False).head()    # in decending order.
```

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']


df_pr.sort_values(by = 'Reviews' , ascending = True)
```

```python
# Perform groupby operation and calculate mean rating for each category
## now result in serirs form.

#            df.groupby('Category').mean()['Rating'].sort_values(ascending = False)
```

```python
## here Category is the primary column and Type is sencondary we get the result 
## on the basis of category by using the values of type.


df.groupby('Category').count()['Type'].sort_values(ascending = False)
```

```python
## it will give me the all free apps of the 

df_pr = df[df['Type'] == 'Free']
df_pr.groupby('Category').count()['Type'].sort_values(ascending = False)
```
